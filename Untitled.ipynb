```python
import requests
import json
import time
```

```python
league_id = 15089
lima_major_matches = requests.get(f'https://api.opendota.com/api/leagues/{league_id}/matches')
```

```python
# download whole data set and filter only the match_id
# using the opendota api to download by league id only gives
# reduced data set about the matches
lima_major = json.loads(lima_major_matches.content)
lima_major_match_ids = [match['match_id'] for match in lima_major]
```

```python
# make sure the number of matches is correct -> 193
len(lima_major_match_ids)
```

```python
lima_major_matches = []
for match_id in lima_major_match_ids: 
    time.sleep(1.05)
    r = requests.get(f'https://api.opendota.com/api/matches/{match_id}')
    match = json.loads(r.content)
    lima_major_matches.append(match)
```

```python
len(lima_major_matches)
```

```python
type(lima_major_matches)
```

```python
matches_dump = json.dumps(lima_major_matches, indent=4)

with open('matches.json', 'w') as outfile:
    outfile.write(matches_dump)
```
